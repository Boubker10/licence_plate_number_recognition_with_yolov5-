```python
import torch
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import easyocr
from google.colab import files

# Télécharger une image
uploaded = files.upload()

def convert_licence_to_text(boxes, img_cv2):
    global licences_text
    reader = easyocr.Reader(['en'])
    licences_text = []
    for (x1, y1, x2, y2) in boxes:
        cropped_image = img_cv2[int(y1):int(y2), int(x1):int(x2)]
        ocr_result = reader.readtext(cropped_image)
        text_result = " ".join([res[1] for res in ocr_result])
        licences_text.append(text_result)

def annotator(results, frame, class_names, Licence=False):
    xyxy = results.xyxy[0].cpu().numpy()
    confidences = results.xyxy[0][:, 4].cpu().numpy()
    class_ids = results.xyxy[0][:, 5].cpu().numpy().astype(int)

    if Licence:
        convert_licence_to_text(xyxy, frame)
        labels = licences_text
    else:
        labels = [
            f"{class_names[class_id]} {confidence:.2f}"
            for class_id, confidence in zip(class_ids, confidences)
        ]

    for i, (x1, y1, x2, y2) in enumerate(xyxy[:, :4]):
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
        label = labels[i]

        # Dessiner les boîtes englobantes
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    return frame

if uploaded:
    img_path = next(iter(uploaded))  # Obtenir le nom du fichier téléchargé

    # Lire l'image avec OpenCV
    img_cv2 = cv2.imread(img_path)
    if img_cv2 is None:
        print(f"Erreur de lecture de l'image à partir de {img_path}")
    else:
        print(f"Image {img_path} lue avec succès")

    # Convertir l'image de BGR (OpenCV) à RGB (PIL)
    img_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)

    # Convertir l'image au format PIL
    img_pil = Image.fromarray(img_rgb)

    # Charger le modèle YOLOv5
    model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

    # Effectuer la détection
    results = model(img_pil)

    # Vérification des résultats de la détection
    if results is None or results.xyxy[0].cpu().numpy().size == 0:
        print("Aucune détection trouvée")
    else:
        print("Détections trouvées")

    # Annoter l'image avec les résultats de détection de YOLOv5
    class_names = model.names
    img_annotated = annotator(results, img_cv2, class_names,licence=True)

    # Convertir l'image de BGR à RGB pour l'affichage
    img_rgb = cv2.cvtColor(img_annotated, cv2.COLOR_BGR2RGB)

    # Afficher les résultats en utilisant matplotlib
    plt.figure(figsize=(10, 10))
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
else:
    print("Aucun fichier n'a été téléchargé.")
```
